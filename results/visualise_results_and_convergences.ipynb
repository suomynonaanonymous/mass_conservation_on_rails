```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import torch
```

# dfNN

```python
model_name = "dfNN"

# load and show for run 1
curves = torch.load(
    f"../trained_models/{model_name}/{model_name}_loss_curves_epochs_3000_run_1.pt", 
    weights_only = False)

plt.figure(figsize = (10, 5))
plt.plot(curves["train_RMSE"], label = 'Train RMSE')
plt.plot(curves["test_RMSE"], label = 'Test RMSE')
plt.xlabel('Epoch')
plt.ylabel('RMSE')
plt.title(f'Training and Test RMSE over Epochs - {model_name}')
plt.ylim(0, 1)
plt.legend()
plt.grid()
plt.show()
```

```python
# from ..configs import NUM_RUNS
dfNN_test_rmses = []
dfNN_test_maes = []
dfNN_test_mads = []

NUM_RUNS = 5

for run in range(NUM_RUNS):
    path = "dfNN" + "/" + "dfNN" + "_metrics_run_" + str(run) + ".csv"
    results_from_run = pd.read_csv(path)
    dfNN_test_rmses.append(results_from_run["RMSE"].item())
    dfNN_test_maes.append(results_from_run["MAE"].item())
    dfNN_test_mads.append(results_from_run["MAD"].item())

print(f"dfNN Test RMSEs: {dfNN_test_rmses}")
print(f"dfNN Mean RMSE: {np.mean(dfNN_test_rmses):.3f}")
print(f"dfNN std RMSE: {np.std(dfNN_test_rmses):.3f}")
print()
print(f"dfNN Test MAEs: {dfNN_test_maes}")
print(f"dfNN Mean MAE: {np.mean(dfNN_test_maes):.3f}")
print(f"dfNN std MAE: {np.std(dfNN_test_maes):.3f}")
print()
print(f"dfNN Test MADs: {dfNN_test_mads}")
print(f"dfNN Mean MAD: {np.mean(dfNN_test_mads):.3f}")
print(f"dfNN std MAD: {np.std(dfNN_test_mads):.3f}")
```
